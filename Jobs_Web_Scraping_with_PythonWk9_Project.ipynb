## <font color='#2F4F4F'>Prerequisites</font>

```
# We first import the required libraries
# ---
#
import pandas as pd             # library for data manupation
import requests                 # library for fetching a web page 
from bs4 import BeautifulSoup   # library for extrating contents from a webpage 
```

## <font color='#2F4F4F'>Step 1: Obtaining our Data</font>

```
# PigiaMe: https://www.pigiame.co.ke/it-software-jobs
# ---
#
pigia_me = requests.get('https://www.pigiame.co.ke/it-software-jobs')
pigia_me
```

```
# MyJobMag: https://www.myjobmag.co.ke/jobs-by-field/information-technology
# ---
#
my_jobmag = requests.get('https://www.myjobmag.co.ke/jobs-by-field/information-technology')
my_jobmag
```

```
# KenyanJob: https://www.kenyajob.com/job-vacancies-search-kenya?f%5B0%5D=im_field_offre_secteur%3A133
# ---
#
kenyan_jobs = requests.get('https://www.kenyajob.com/job-vacancies-search-kenya?f%5B0%5D=im_field_offre_secteur%3A133')
kenyan_jobs
```

## <font color='#2F4F4F'>Step 2: Parsing</font>

```
# Parsing our document: pigia_me
# ---
# 
pigia_soup = BeautifulSoup(pigia_me.text, "html.parser")
```

```
print(pigia_soup.prettify())
```

```
# Parsing our document: my_job_mag
# ---
#  
myjob_soup = BeautifulSoup(my_jobmag.text, "html.parser")
print(myjob_soup.prettify())
```

```
# Parsing our document: kenyan_job
# ---
# 
kenjob_soup = BeautifulSoup(kenyan_jobs.text, "html.parser")
print(kenjob_soup.prettify())
```

## <font color='#2F4F4F'>Step 3: Extracting Required Elements</font>

```
# 1. Extracting job titles and links: pigia me
# ---
# 

pigiame_results = pigia_soup.find_all('div', {'class': 'listings-cards__list-item'})
pigiame_results
```

```
pigiame_links = []
pigiame_titles = []

for results in pigiame_results:
  links = results.find_all('a')  # links
  for link in links:
    urls = link.get('href')
    pigiame_links.append(urls)  
  titles = results.find_all('div', {'class': 'listing-card__header__title'})
  for title in titles:
    text = title.get_text().strip() # titles
    pigiame_titles.append(text)
```

```
df1 = pd.DataFrame({"link_content": pigiame_titles, "link_url": pigiame_links})
df1.head(10)
```

```
pigiame_links
```

```
# 2. Extracting job titles: my_job_mag
# ---
# 
myjobmag_results = myjob_soup.find_all('h2')
myjobmag_results
```

```
len(myjobmag_results)
```

```
p_tags_text = [tag.get_text().strip() for tag in myjobmag_results]
p_tags_text
```

```
#
link_content = []
link_url = []

# Getting all our <h2> tags 
# ---
#
myjobmag_results = myjob_soup.find_all('h2')

# We the loop through these tags
for results in myjobmag_results:
   
    # Getting our text from each tag
    text = results.get_text()

    # We concatenate our domain with href link that we scrape
    # in order to form a full link
    
    link = 'https://www.myjobmag.co.ke/jobs-by-field/information-technology' #+ results.get('href')

    # Then appending the text to our link_content list
    link_content.append(text)

    # Then appending the text to our link_url list
    link_url.append(link)
```

```
df2 = pd.DataFrame({"link_content": link_content, "link_url": link_url})
df2.head(10)
```

```
# 3. Extracting job titles: kenya_job
# ---
#
kenjob_results = kenjob_soup.find_all('h5')
kenjob_results
```

```
p_tags_text = [tag.get_text().strip() for tag in kenjob_results]
p_tags_text
```

```
#
link_content = []
link_url = []

# Getting all our <a> tags 
# ---
#
kenjob_results = kenjob_soup.find_all('h5')

# We the loop through these tags
for results in kenjob_results:
   
    # Getting our text from each tag
    text = results.get_text()

    # We concatenate our domain with href link that we scrape
    # in order to form a full link
    link = 'https://www.kenyajob.com/job-vacancies-search-kenya?f%5B0%5D=im_field_offre_secteur%3A133' #+ results.get('href')

    # Then appending the text to our link_content list
    link_content.append(text)

    # Then appending the text to our link_url list
    link_url.append(link)
```

```
df3 = pd.DataFrame({"link_content": link_content, "link_url": link_url})
df3.head(10)
```

## <font color='#2F4F4F'>Step 4: Saving our Data</font>

```
# Saving the scraped contents in a dataframe and preview our data
# ---
#
frames = [df1, df2, df3]
 
final_df = pd.concat(frames)
final_df
```
